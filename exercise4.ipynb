```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib import patheffects

# Create a figure and axis with Cartopy projection
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': ccrs.PlateCarree()})

# Set the extent of the map (this will ensure the map covers the entire globe)
ax.set_extent([-180, 180, -90, 90])

# Add coastlines to the map
ax.coastlines()

# Add an arrow pointing to Charlotte, North Carolina
charlotte_lon, charlotte_lat = -80.8431, 35.2271  # Charlotte's coordinates

# Add an arrow
ax.annotate('', xy=(charlotte_lon, charlotte_lat), xytext=(-80, 40),
            arrowprops=dict(facecolor='red', edgecolor='red', arrowstyle='->', lw=2))

# Add a label with the text "We are here"
ax.text(charlotte_lon + 0.5, charlotte_lat, "We are here", fontsize=12, color='red', weight='bold',
        path_effects=[patheffects.withStroke(linewidth=3, foreground="white")])

# Add labels for major oceans (adjusted Pacific Ocean position further to the right)
ax.text(-45, 10, 'Atlantic Ocean', fontsize=12, color='blue', weight='bold',
        ha='center', va='center')
ax.text(170, 0, 'Pacific Ocean', fontsize=12, color='blue', weight='bold',  # Moved further right
        ha='center', va='center')
ax.text(80, -30, 'Indian Ocean', fontsize=12, color='blue', weight='bold',
        ha='center', va='center')
ax.text(0, 60, 'Arctic Ocean', fontsize=12, color='blue', weight='bold',
        ha='center', va='center')

# Add a title
plt.title("World Map with Ocean Labels and Arrow to Charlotte, North Carolina")

# Show the map
plt.show()
```
